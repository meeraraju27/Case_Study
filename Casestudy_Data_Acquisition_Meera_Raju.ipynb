```python
import pandas as pd
```

###Extract Data:


```python
# 1. GDP Excel
df_gdp  = pd.read_excel("/content/gdp.xlsx")
df_gdp
```

```python
# 2. Internet users JSON
df_internet = pd.read_json("/content/internet_users.json")
df_internet
```

```python
#3.Literacy XML
df_literacy = pd.read_xml("/content/literacy_rate.xml")
df_literacy
```

```python
#4.Population CSV
df_population = pd.read_csv("/content/population.csv")
df_population
```

###Transform Data:


```python
#Standardize country names across all datasets
df_gdp = df_gdp.rename(columns={'Country': 'country'})
df_gdp
```

```python
#Standardize country names across all datasets
df_literacy = df_literacy.rename(columns={'name': 'country'})
df_literacy
```

```python
#Convert numeric columns to appropriate types
#check the datatypes
print("Population:\n", df_population.dtypes)
print("\nInternet Users:\n", df_internet.dtypes)
print("\nLiteracy Rate:\n", df_literacy.dtypes)
print("\nGDP:\n", df_gdp.dtypes)
```

```python
# Handle missing or erroneous data
print(df_population.isnull().sum())
print("\n",df_internet.isnull().sum())
print("\n",df_literacy.isnull().sum())
print("\n",df_gdp.isnull().sum())
```

```python
#Create new columns such as internet penetration rate (internet users / population × 100)
df_internet['internet_penetration_rate'] = (df_internet['internet_users'] / df_population['population']) * 100
df_internet
```

###Integrate Data:

```python
#Merge the datasets based on the country name

# Merge population and internet users
merged_data = pd.merge(df_population, df_internet, on='country', how='inner')

# Merge with GDP
merged_data = pd.merge(merged_data, df_gdp, on='country', how='inner')

# Merge with Literacy
merged_data = pd.merge(merged_data, df_literacy, on='country', how='inner')

#final merged dataset
print(" Merged dataset created with", len(merged_data), "rows")
merged_data
```

```python
#Handle unmatched or missing records
missing_data = merged_data.isnull().sum()
missing_data
```

```python
#Validate the data for consistency and completeness
merged_data.info()
```

```python
#Validate the data for consistency and completeness
invalid_literacy = merged_data[(merged_data['literacy_rate'] < 0) | (merged_data['literacy_rate'] > 100)]
invalid_literacy

invalid_population = merged_data[merged_data['population'] <= 0]
invalid_population
```

###Analyze Data:


```python
#Find countries with highest internet penetration rates
highest_penetration_countries = merged_data.sort_values(by='internet_penetration_rate', ascending=False)
highest_penetration_countries.head(10)
```

```python
#Calculate average literacy rate across countries
avg_literacy_rate = merged_data['literacy_rate'].mean()
print("Average literacy rate across countries:", avg_literacy_rate)
```

```python
#Investigate relationships between GDP, population, literacy, and internet usage
correlation_matrix = merged_data[['GDP', 'population', 'literacy_rate', 'internet_users']].corr()
correlation_matrix
```

###Load into SQLite:

```python
#Create an SQLite database
import sqlite3
conn = sqlite3.connect("merged_data")
cursor = conn.cursor()
```

```python
#Load the final merged dataset into a database table
merged_data.to_sql("merged_data", conn, if_exists="replace", index=False)
```

```python
cursor.execute("SELECT * FROM merged_data LIMIT 5")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

###● Write and execute SQL queries to extract insights

```python
#a. Countries with highest internet penetration
cursor.execute("""SELECT country, internet_penetration_rate
                FROM merged_data ORDER BY internet_penetration_rate
                DESC LIMIT 10""")
# Fetch results
top10_penetration = cursor.fetchall()
for row in top10_penetration:
    print(row)
```

```python
#b. Average literacy rate
cursor.execute("""SELECT AVG(literacy_rate)
                FROM merged_data""")
avg_literacy = cursor.fetchone()
print("Average literacy rate:", avg_literacy )
```

```python
#c.Correlation between literacy rate and internet penetration(pandas)
correlation = merged_data['literacy_rate'].corr(merged_data['internet_penetration_rate'])
print("Correlation (literacy vs internet penetration):", correlation)
```

```python
#c.Correlation between literacy rate and internet penetration
cursor.execute("""SELECT
    (AVG( internet_penetration_rate	* literacy_rate) - AVG(internet_penetration_rate	) * AVG(literacy_rate)) /
    (SQRT((AVG(internet_penetration_rate	 * internet_penetration_rate	) - AVG(internet_penetration_rate	)*AVG(internet_penetration_rate	)) *
          (AVG(literacy_rate * literacy_rate) - AVG(literacy_rate)*AVG(literacy_rate))))
    AS correlation
FROM merged_data;""")

correlation = cursor.fetchone()
print("Correlation (literacy vs internet penetration):", correlation)
```

```python
#d. GDP per capita analysis
cursor.execute("""SELECT country,
           gdp * 1.0 / population AS gdp_per_capita
           FROM merged_data;""")

gdp_per_capita = cursor.fetchall()
for row in gdp_per_capita:
    print(row)
```

```python
#e. Which countries have a GDP per capita above $10,000?
cursor.execute("""SELECT country,
           gdp * 1.0 / population AS gdp_per_capita
           FROM merged_data
           WHERE gdp_per_capita > 10000;""")
gdp_per_capita = cursor.fetchall()
for row in gdp_per_capita:
    print(row)
```

```python
#f. What is the total population covered in the dataset?
cursor.execute("""SELECT SUM(population) AS total_population
           FROM merged_data;""")
total_population = cursor.fetchone()
print("Total population covered in the dataset:", total_population)
```

```python
#g. Which countries have the lowest literacy rates, and how does that impact internet access?
cursor.execute("""SELECT country, literacy_rate, internet_penetration_rate
           FROM merged_data
           ORDER BY literacy_rate ASC
           LIMIT 10;""")
lowest_literacy = cursor.fetchall()
for row in lowest_literacy:
    print("country :",row[0],"    literacy_rate :",row[1],"   internet_penetration_rate :",row[2])
```

```python
#h. What are the top 5 wealthiest countries by total GDP, and how does that compare with population size?
cursor.execute("""SELECT country, gdp, population
           FROM merged_data
           ORDER BY gdp DESC
           LIMIT 5;""")
top_wealthy = cursor.fetchall()
for row in top_wealthy:
    print(row)


```

```python
#i. Find countries where internet users exceed 70% of the population
cursor.execute("""SELECT country, internet_users, population
           FROM merged_data
           WHERE internet_users > 0.7 * population;""")
high_internet_users = cursor.fetchall()
for row in high_internet_users:
    print("country:",row[0],"   internet_users:",row[1],"   population:",row[2])
```

```python
#j. What is the average GDP per capita for countries with internet penetration above 50%?
cursor.execute("""SELECT AVG(gdp * 1.0 / population) AS avg_gdp_per_capita
           FROM merged_data
           WHERE internet_penetration_rate > 50;""")
avg_gdp_per_capita = cursor.fetchone()
print("Average GDP per capita for countries with internet penetration above 50%:", avg_gdp_per_capita)
```

```python
#How many countries have a literacy rate above 90%, and what is their average internet penetration?
cursor.execute("""SELECT COUNT(*) AS country_count, AVG(internet_penetration_rate) AS avg_internet_penetration
           FROM merged_data
           WHERE literacy_rate > 90;""")
result = cursor.fetchone()
print("Number of countries with literacy rate above 90%: ", result[0])
print("Average internet penetration for countries with literacy rate above 90%: ", result[1])
```
