# 2. Weather Data Analyzer
a. Analyze weather data (temperature, humidity, etc.) for a given region

b. Use variables and data types to store weather data

c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)

d. Use functions to calculate average temperature and humidity, and print
results

```python
def average_temperature(temperatures):
  return sum(temperatures)/len(temperatures)

def average_humidity(humiditys):
  return sum(humiditys)/len(humiditys)

def determine_weather_condition(avg_temp, avg_humidity):
  if avg_temp > 30 and avg_humidity < 50:
    return "It's a Sunny week"
  elif avg_temp < 20 and avg_humidity > 70:
    return "It's a Rainy week"
  elif 20 <= avg_temp <= 30 and 50 <= avg_humidity <= 70:
    return " Pleasant"
  else:
    return "Mixed Conditions"

print("welcome to weather data analyzer")
print("Enter the weather data of 1 week:")
temperatures = []
humiditys = []
for i in range(1,8):
  print(f"for day {i}:")
  temp = int(input("Enter the temperature:"))
  humid = int(input("Enter the humidity:"))
  temperatures.append(temp)
  humiditys.append(humid)
avg_temp = average_temperature(temperatures)
avg_humid = average_humidity(humiditys)
condition = determine_weather_condition(avg_temp, avg_humid)

print("\n****** Weather Analysis Result******")
print(f"Average Temperature: {avg_temp:.2f}°C")
print(f"Average Humidity: {avg_humid:.2f}%")
print(f"Overall Condition: {condition}")
```
