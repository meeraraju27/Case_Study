##Numpy

###1. Basic Array Operations

Convert the mpg column into a NumPy array and calculate:
○ The mean, median, and standard deviation of mpg.
○ The number of cars with mpg greater than 25


```python
mpg_array = df['mpg'].to_numpy()

mpg_mean = np.mean(mpg_array)
mpg_median = np.median(mpg_array)
mpg_std = np.std(mpg_array)

print(f"Mean MPG: {mpg_mean}")
print(f"Median MPG: {mpg_median}")
print(f"Std deviation:{mpg_std}")

no_of_cars = np.sum(mpg_array > 25)
print(f"Number of cars with mpg greater than 25: {no_of_cars}")
```

###2. Filtering
Using NumPy, filter all cars with more than 6 cylinders.

###3. Statistical Analysis
Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy.

```python
weight_array = df['weight'].to_numpy()
weight_percentiles = np.percentile(weight_array, [25, 50, 75])
print("25th, 50th, and 75th percentiles = ",weight_percentiles)
```

###4. Array Manipulation
Convert the acceleration column into a NumPy array and normalize its values
(scale between 0 and 1).

```python
acceleration_array = df['acceleration'].to_numpy()
acceleration_min = np.min(acceleration_array)
acceleration_max = np.max(acceleration_array)
normalized_acceleration = (acceleration_array - acceleration_min) / (acceleration_max - acceleration_min)
print(normalized_acceleration)
```

Increase all horsepower values by 10% and store the updated values in a new

###10.Conditional Aggregates
Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

###10. Finding Anomalies
Identify potential outliers in the mpg column using the Interquartile Range (IQR)
method. Specifically:
● Calculate the IQR for mpg.
2
● Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 +
1.5 * IQR.
● Create a DataFrame of cars classified as outliers, displaying car_name, mpg,
and model_year.

```python
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_df = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]
outliers_df = outliers_df[['car name','mpg','model year']]
outliers_df
```
